```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean, std_dev = 0, 1
```

```python
def probability_dist_function(x: float, mean: float, std_dev: float) -> float:
  """Probability distribution function of the normal distribution"""
  p = (1/np.sqrt(2*np.pi))*np.exp(-(x - mean)**2/(2*std_dev**2))

  return p
```

```python
start_x, stop_x, num = -5, 5, 200
x_axis = np.linspace(start=start_x, stop=stop_x, num=num)
y_axis = probability_dist_function(x=x_axis, mean=mean, std_dev=std_dev)
```

```python
plt.plot(x_axis, y_axis, color='r')
plt.title('Normal Distribution Curve')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.show()
```

```python
sample_size = 100
sample = np.random.normal(mean, std_dev, sample_size)
sample
```

```python
plt.hist(sample)
plt.title('Normally distributed sample')
plt.xlabel('x value')
plt.ylabel('n observations')
plt.show()
```

```python
# estimate the mean and the std_dev parameters from the sample:
sample_mean = np.mean(sample)
sample_std_dev = np.std(sample)

sample_mean, sample_std_dev
```

```python
# estimated probability distribution function from the estimated parameters
y_axis_estimated = probability_dist_function(x=x_axis, mean=sample_mean, std_dev=sample_std_dev)

plt.hist(sample, density=True)
plt.plot(x_axis, y_axis, color='r', label='true_pdf')
plt.plot(x_axis, y_axis_estimated, color='b', label='estimated_pdf')
plt.title('Estimated p.d.f vs the true p.d.f')
plt.xlabel('x value')
plt.ylabel('p(x)')
plt.legend()
plt.show()
```
